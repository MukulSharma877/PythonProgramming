## Problem Statement 1

```python
# Load the 'Breast_Cancer_Dataset.csv' dataset into a DataFrame and perform the following tasks
```

```python
import pandas as pd
import numpy as np
```

```python
cancer = pd.read_csv("Breast_Cancer_Dataset.csv")
cancer.head()
```

```python
# Identify the null values and remove the null rows and columns by using the dropna() function

print(cancer.isnull().sum())
cancer = cancer.dropna(axis=1)
```

```python
# Encode the 'diagnosis' column using the LabelEncoder()

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
cancer["diagnosis"] = le.fit_transform(cancer["diagnosis"])
cancer.head()
```

```python
# Considering the 'diagnosis' column as the target, separate the target variable and the feature vectors

X = cancer.drop("diagnosis", axis = 1)
y = cancer["diagnosis"]

print(X.head())
print(y.head())
```

```python
# Split the dataset into the training set and test set in a 70:30 ratio

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Building a Logistic Regression, Naive Bayes, Decision Tree (CART), K-NN, SVM, and RandomForestClassifier models
# Also print their accuracies

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score


# Logistic Regression
lr = LogisticRegression()
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)
print("Logistic Regression -> ACC: %", accuracy_score(y_test, lr_pred) * 100)


# Naive Bayes
nb = GaussianNB()
nb.fit(X_train, y_train)
nb_pred = nb.predict(X_test)
print("Naive Bayes -> ACC: %", accuracy_score(y_test, nb_pred) * 100)


# Decision Tree (CART)
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)
print("Decision Tree (CART) -> ACC: %", accuracy_score(y_test, dt_pred) * 100)


# K-NN
knn = KNeighborsClassifier(n_neighbors=19)
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)
print("K-NN -> ACC: %", accuracy_score(y_test, knn_pred) * 100)


#SVM
svm = SVC()
svm.fit(X_train, y_train)
svm_pred = svm.predict(X_test)
print("SVM -> ACC: %", accuracy_score(y_test, svm_pred) * 100)


# Random Forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
print("RandomForestClassifier -> ACC: %", accuracy_score(y_test, rf_pred) * 100)
```

```python
# Calculate and plot the confusion matrix

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

cf_matrix = confusion_matrix(y_test, dt_pred)
print(cf_matrix)
sns.heatmap(cf_matrix, cmap="Blues", annot=True)
plt.xlabel("Predicted Values")
plt.ylabel("True Values")
plt.show()
```

```python
# Calculate the  ROC_AUC score based on the False Positive Rate (FPR) and True Positive Rate (TPR)

from sklearn.metrics import roc_auc_score

print("roc_auc score is: \n", roc_auc_score(y_test, dt_pred))
```

```python
# Plot the ROC Curve using the Matplotlib library

y_prob = rf.predict_proba(X_test)

def calculate_tpr_fpr(y_real, y_pred):
    cm = confusion_matrix(y_real, y_pred)
    TN = cm[0, 0]
    FP = cm[0, 1]
    FN = cm[1, 0]
    TP = cm[1, 1]
    tpr =  TP/(TP + FN)
    fpr = 1 - TN/(TN+FP)
    return tpr, fpr

def get_n_roc_coordinates(y_real, y_prob, n = 5):
    tpr_list = [0]
    fpr_list = [0]
    for i in range(n):
        threshold = i/n
        y_pred = y_prob[:, 1] > threshold
        tpr, fpr = calculate_tpr_fpr(y_real, y_pred)
        tpr_list.append(tpr)
        fpr_list.append(fpr)
    return tpr_list, fpr_list

def plot_roc_curve(tpr, fpr, scatter = True):
    plt.figure(figsize = (5, 5))
    if scatter:
        sns.scatterplot(x = fpr, y = tpr)
    sns.lineplot(x = fpr, y = tpr)
    sns.lineplot(x = [0, 1], y = [0, 1], color = 'green')
    plt.xlim(-0.05, 1.05)
    plt.ylim(-0.05, 1.05)
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    
tpr, fpr = get_n_roc_coordinates(y_test, y_prob)
plot_roc_curve(tpr, fpr)
```

## Problem Statement 3

```python
# Load the ‘voice.csv’ dataset into a DataFrame and perform the following tasks
```

```python
voice = pd.read_csv("voice.csv")
voice.head()
```

```python
# Considering the ‘label’ column as the target variable, rename the column as ‘Gender_Identified’

voice.rename({"label" : "Gender_Identified"}, axis = 1, inplace=True)
voice.head()
```

```python
# Using the preprocessing() function, label the target column

from sklearn.preprocessing import LabelEncoder

def preprocessing(x):
    le = LabelEncoder()
    x["Gender_Identified"] = le.fit_transform(x["Gender_Identified"])
    return x

voice = preprocessing(voice)
voice.head()
```

```python
# Separate the target variable and the feature vectors

X = voice.drop("Gender_Identified", axis = 1)
y = voice["Gender_Identified"]

print(X.head(), y.head(), sep = "\n")
```

```python
# Build a RandomForestClassifiermodel and find the best parameters using a Grid search

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

params = { "criterion" : ["gini", "entropy"], "n_estimators": [100, 150, 200,300]}
rf=GridSearchCV(estimator=RandomForestClassifier(),param_grid=params,cv=3,scoring='accuracy')
rf.fit(X,y)

pd.DataFrame(rf.cv_results_).sort_values("rank_test_score")
```

```python
# Print the best parameters and the best estimator

print(rf.best_params_)

print(rf.best_estimator_)
```
